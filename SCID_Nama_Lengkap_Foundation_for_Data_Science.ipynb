###### 2.

<p>Hitunglah BMI setiap siswa dengan rumus BMI = Berat badan (kg) / (Tinggi Badan (m))^2</p>

````python
# Bagian 2
def bmi(berat_badan, tinggi_badan):
  return berat_badan / (tinggi_badan / 100)**2
````

###### 3.

<p>Buatlah program untuk menghitung BMI!</p>

````python
# Bagian 2
hasil_bmi = pd.Series(map(bmi, list(df_19["Berat Badan (Kg)"]), list(df_19["Tinggi Badan (Cm)"])))
display(hasil_bmi)
````

###### 4.

<p>Update dataframe df dengan menambahkan hasil perhitungan soal nomor 2, beri nama kolom BMI.</p>

````python
# Bagian 2
df_19["BMI"] = hasil_bmi
display(df_19)
````

###### 5.

<p>Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan nama kolom Status BMI.</p>

<ul>
  <li>Normal: 18.5-24.9</li>
  <li>Overweight: 25.0-29.9</li>
  <li>Obesity: 30.0-34.9</li>
  <li>Underweight: < 18.5</li>
</ul>

````python
# Bagian 3
def status_bmi(bmi):
  if bmi >= 18.5 and bmi <= 24.9:
    return "Normal"
  if bmi >= 25.0 and bmi <= 29.9:
    return "Overweight"
  if bmi >= 30.0 and bmi <= 34.9:
    return "Obesity"
  if bmi < 18.5:
    return "Underweight"
  return ""

hasil_status_bmi = pd.Series(map(status_bmi, list(df_19["BMI"])))

df_19["Status BMI"] = hasil_status_bmi
display(df_19)
````

### SOAL 2

<p>Perusahaan Investasi Jamin-Sugih Co kami memberikan penawaran investasi terbaik dengan pembagian keuntungan tetap sebesar 25% setiap tahun.</p>

<p>Buatlah program Python yang dapat menerima input dari pengguna dalam bentuk:</p>
<ul>
  <li>Modal Investasi (bilangan bulat antara 0-1.000.000.000)</li>
  <li>Jangka Waktu Investasi (bilangan bulat antara 1-25 dalam tahun)</li>
  <li>Selain itu, program akan menampilkan error message bahwa modal atau tenure tidak sesuai ketentuan.</li>
</ul>

<p>Program akan menampilkan informasi akumulasi investasi setiap tahun, menggunakan iterasi untuk setiap tahun.<p>

<p>Contoh:</p>

<p>Input</p>
<ul>
  <li>Money Capital : 100000</li>
  <li>Investment Tenure (years) : 2</li>
</ul>

<p>Output</p>
<ul>
  <li>Year 1, your total investment : 125000.0</li>
  <li>Year 2, your total investment : 156250.0</li>
</ul>

````python
# Bagian 5
print("Input")
capital_gain = 0.25
money_capital = int(input(">> Money capital : "))
investment_tenure = int(input(">> Investment tenure (years) : "))

print("Output")
if (money_capital >= 0 and money_capital <= 1_000_000_000) and (investment_tenure >= 1 and investment_tenure <= 25):
  total_investment = money_capital
  for i in range(investment_tenure):
    total_investment *= (1 + capital_gain)
    print(">> Year {}, your total investment : {}".format(i + 1, total_investment))
else:
  print("Modal atau tenure tidak sesuai ketentuan")
````
